```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
df=pd.read_csv('Amazon Sales data.csv')
df
```

```python
df=df.isnull().sum()
df
```

```python
df.info()
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.nunique()
df
```

```python
df.drop(columns=['Order ID'],inplace=True)
df
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df
```

```python
df['year']=df['Order Date'].dt.year
df['month']=df['Order Date'].dt.month 
df['day']=df['Order Date'].dt.day
df['day_name']=df['Order Date'].dt.day_name()
df['dayofweek']=df['Order Date'].dt.dayofweek
df['quarter']=df['Order Date'].dt.quarter
df['lead time']=(df['Ship Date']-df['Order Date'])
df
```

```python
df['lead time'] = df['lead time'].dt.days
df
```

```python

cd_2010['Unit Price'] = cd_2010['Unit Price'].round(2)
cd_2010['Unit Cost'] = cd_2010['Unit Cost'].round(2)
item_cost_price_2010 = cd_2010.groupby('Item Type', as_index=False)[['Unit Price', 'Unit Cost']].mean()

icp_2010 = make_subplots(specs=[[{"secondary_y": True}]])

icp_2010.add_trace(
    go.Scatter(
        x=item_cost_price_2010['Item Type'],
        y=item_cost_price_2010['Unit Price'],
        mode='lines+markers+text',  # Adding text on the markers
        name='Unit Price',
        line=dict(color='blue'),
        text=item_cost_price_2010['Unit Price'].round(0).astype(int),  # Display Unit Price as text
        textposition='top right',
        textfont=dict(color='blue', size=12), 
    ),
    secondary_y=False
)

icp_2010.add_trace(
    go.Scatter(
        x=item_cost_price_2010['Item Type'],
        y=item_cost_price_2010['Unit Cost'],
        mode='lines+markers+text',  # Adding text on the markers
        name='Unit Cost',
        line=dict(color='green'),
        text=item_cost_price_2010['Unit Cost'].round(0).astype(int),# Display Unit Cost as text
        textposition='bottom left',
        textfont=dict(color='green', size=12), # Position of text
    ),
    secondary_y=True
)

icp_2010.update_layout(
    title="Unit Price V/S  Unit Cost by Item Type [2010]",
    xaxis_title="Item Type",
    yaxis_title="Unit Price($)",
    legend_title="Metrics",
    template="plotly_white"
)

icp_2010.update_layout(
    yaxis2=dict(title="Unit Cost($)")
)

icp_2010.show()
```
